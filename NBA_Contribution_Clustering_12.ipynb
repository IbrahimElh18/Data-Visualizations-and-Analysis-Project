## **Techniques**  
---
<span style="color:#189AB4">**K-Means Clustering**</span> -   

K-means clustering is a popular unsupervised machine learning algorithm used for grouping data points into distinct clusters 
based on their similarities. The algorithm aims to partition the data into K clusters, where K represents the predetermined number of clusters desired.  
In the context of the dataset, K-means clustering can be used to identify patterns or groupings among the basketball players based on their   
performance attributes such as games played (G), minutes played (MP), total points scored (PTS), total rebounds (TRB), assists (AST), steals (STL), blocks (BLK), turnovers (TOV), and personal fouls (PF).  

By applying K-means clustering to this dataset, the algorithm will automatically assign players to different clusters based on their performance attributes.   
This can help in uncovering insights such as identifying groups of players with similar performance profiles or discovering any distinct player archetypes within the   
LA Lakers team during the 2021-2022 season.  

The clustering results can assist in various analyses and decision-making processes. For example, it can aid in player evaluation,   
identifying key contributors, understanding player roles, or even forming game strategies based on the strengths and weaknesses of different player clusters.  

--- 
<span style="color:#189AB4">**Chi-Squared**</span> -  


The chi-squared test is a test used to determine whether there is a significant association between two different variables or in this case, features.     
By performing the chi 2 test on this dataset after we cleaned and filtered our data from noisy or extraneous data, we can see whether there is a significant relationship between the different features to points scored specified by any variable/feature. In this we would want to compare points scored to anything that can relate to a player scoring and  
compare whether there is a strong correlation between a player scoring a point vs the compared feature. 

By understanding the results of the chi 2 test, by finding the chi2 value and it's P value, we compare it's calculated P value to the significance value (usually 0.05).  
If the Calculated P value is greater than the Significance value we fail to reject the null hypothesis and stat that it has no association. On the otherhand, if the   
calculated P value is less than the Significance value, we reject the null hypothesis and conclude that there is association between the two features.

---
<span style="color:#189AB4">**Feature Selection**</span> -    

Feature selection is the process of identifying and selecting a subset of relevant features or variables from a larger set of available features. It aims to improve model  
performance, reduce complexity, and enhance interpretability by focusing on the most informative and influential features.  

In the dataset, which contains various player performance attributes -- or features in this matter, feature selection can be used to identify the most influential  
and significant features for predicting a particular outcome or analyzing the data.   

By applying feature selection techniques, we can determine which features have the strongest relationships with the target variable or provide the most valuable information  
for the analysis at hand. This process helps to streamline the dataset and focus on the more meaningful features, reducing noise and improving model efficiency  
and interpretability.

## **Chi squared test**

```python
#Contingency table for each comparison
comparisons = [('AST', 'PTS'), ('TRB', 'PTS'), ('STL', 'PTS'), ('BLK', 'PTS')]

for var1, var2 in comparisons:
    contingency_table = pd.crosstab(filtered_data[var1], filtered_data[var2])
    
    #Chi analysis
    chi2, p_value, _, _ = chi2_contingency(contingency_table)
    
    print(f"Chi-square analysis between {var1} and {var2}:")
    print(f"Chi-square statistic: {chi2}")
    print(f"P-value: {p_value}")
    
    if p_value < 0.05:
        print("The p-value is less than 0.05, so there is a significant relationship.")
    else:
        print("The p-value is greater than or equal to 0.05, so there is no significant relationship.")
    
    print()
```

# K Means Clustering

```python
data.describe()
#We see too much variation here, so we need to standardize
```

```python
# standardizing the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# statistics of scaled data
pd.DataFrame(data_scaled).describe()
```

```python
# defining the kmeans function with initialization as k-means++
kmeans = KMeans(n_clusters=2, init='k-means++', n_init='auto', random_state=3)

# fitting the k means algorithm on scaled data
kmeans.fit(data_scaled)
```

```python
# inertia on the fitted data
print("Inertia:", kmeans.inertia_)
```

```python
# fitting multiple k-means algorithms and storing the values in an empty list
SSE = []
for cluster in range(1,20):
    kmeans = KMeans(n_clusters = cluster, init='k-means++', n_init='auto', random_state=3)
    kmeans.fit(data_scaled)
    SSE.append(kmeans.inertia_)

# converting the results into a dataframe and plotting them
frame = pd.DataFrame({'Cluster':range(1,20), 'SSE':SSE})
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
```

```python
# k means using 6 clusters and k-means++ initialization
kmeans = KMeans(n_clusters = 6, init='k-means++', n_init='auto', random_state=3)
kmeans.fit(data_scaled)
pred = kmeans.predict(data_scaled)
```

```python
frame = pd.DataFrame(data_scaled)
frame['cluster'] = pred
frame['cluster'].value_counts()
```

# Feature Selection  

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest, f_regression

# Read the CSV file into a DataFrame
nba_data = pd.read_csv('NBA_Player_Stats.csv')

# Filter the data for the years '2021-2022' and the LA Lakers ('LAL')
filtered_data = nba_data[(nba_data['Year'].str.contains('2021-2022')) & (nba_data['Tm'].str.contains('LAL'))]

# Load the dataset
df = filtered_data

# # Check the columns in the DataFrame
# print(df.columns)

# Extract the feature columns (excluding the target variable)
X = df.select_dtypes(include=[np.number]).drop(['PTS'], axis=1)

# Extract the target variable
y = df['PTS']

# Calculate the correlation between features and target variable
correlation_scores = X.apply(lambda x: x.corr(y))

# Select the top K features based on correlation scores
k = 5
top_k_features = correlation_scores.abs().nlargest(k).index

# Select the corresponding columns from the dataset
X_selected = X[top_k_features]

# Perform univariate feature selection
selector = SelectKBest(score_func=f_regression, k=k)
X_selected = selector.fit_transform(X_selected, y)

# Get the selected feature indices
selected_feature_indices = selector.get_support(indices=True)

# Get the names of the selected features
selected_features = X.columns[selected_feature_indices]

# Print the selected features
print("Selected Features:")
for feature in selected_features:
    print(feature)
```

```python
# Create a new DataFrame with the selected features
selected_data = df[selected_features]

# Display the selected features and corresponding data
print("Selected Features and Data:")
print(selected_data)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer


# Read the CSV file into a DataFrame
nba_data = pd.read_csv('NBA_Player_Stats.csv')

# Filter the data for the years '2021-2022' and the LA Lakers ('LAL')
filtered_data = nba_data[(nba_data['Year'].str.contains('2021-2022')) & (nba_data['Tm'].str.contains('LAL'))]

# Load the dataset
df = filtered_data

# Extract the feature columns (excluding the target variable)
X = df.select_dtypes(include=[np.number]).drop(['PTS'], axis=1)

# Extract the target variable
y = df['PTS']


imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Fit a random forest regression model
model = RandomForestRegressor()
model.fit(X_imputed, y)

# Get feature importances
feature_importances = model.feature_importances_

# Create a DataFrame with feature names and importances
feature_scores = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importances})

# Sort features by their importance scores
feature_scores = feature_scores.sort_values('Importance', ascending=False)

# Print the feature importances
print("Feature Importance:")
print(feature_scores)

# Plot the feature importances
plt.figure(figsize=(10, 6))
plt.barh(feature_scores['Feature'], feature_scores['Importance'])
plt.xlabel('Feature Importance')
plt.ylabel('Features')
plt.title('Feature Importance')
plt.show()
```

First, we used correlation scores to identify the top 5 features that have the highest correlation with the target variable. The selected features are: Rank(Rk), Age, Games Played(G), Games Started(GS), Minutes Played(MP). Features were chosen based on their correlation with the target variable, indicating their potential influence on scoring and game success. Next, we applied univariate feature selection using the SelectKBest method with the f_regression scoring function. Then, we found that the selected the top 5 features based on their scores. The selected features are: Field Goals(FG), Minutes Played(MP), Field Goal Attempts(FGA), 2-Point Attempts(2PA), Total Rebounds(TRB). Comparing the two sets of selected features, we see that the feature "MP" (Minutes Played) appears in both selections, indicating its importance in predicting the win. 

We also took the feature importance, and ranked the features in the following order: Field Goals(FG), Minutes Played(MP), Field Goal Attempts(FGA), 2-Point Attempts(2PA), Total Rebounds(TRB). These results are in line with the features selected through correlation and univariate feature selection, indicating their significance in contributing to scoring and the game win.  

The selected features provide insights into the potentially influential factors in predicting the target variable (PTS) for NBA players on the LA Lakers team during the 2021-2022 season.
